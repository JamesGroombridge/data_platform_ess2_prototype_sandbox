Get list of google drive folder

```python
from google.oauth2 import service_account
from googleapiclient.http import MediaFileUpload
from googleapiclient.discovery import build
from googleapiclient.http import MediaIoBaseUpload
import json

json_key_str = dbutils.secrets.get(scope="apikle_shared_drive", key="info")
info = json.loads(json_key_str)

#Authenticate and build service
SCOPES = ['https://www.googleapis.com/auth/drive']
PARENT_FOLDER_ID = '17NnUwQYh6lvHrk41Cbvju-PgtqwTejKu'

creds = service_account.Credentials.from_service_account_info(info, scopes=SCOPES)
service = build('drive', 'v3', credentials=creds)

page_token = None
print(f"{'Folder Name':<30} | {'Folder ID'}")
print("-" * 60)

while True:
    # Filter for folders only and skip trashed ones
    query = "mimeType = 'application/vnd.google-apps.folder' and trashed = false"
    
    results = service.files().list(
    q=query,
    spaces='drive',
    fields='nextPageToken, files(id, name)',
    pageToken=page_token,
    supportsAllDrives=True,        # Required for Shared Drives
    includeItemsFromAllDrives=True # Required for Shared Drives
).execute()

    for folder in results.get('files', []):
        print(f"{folder['name']:<30} | {folder['id']}")

    page_token = results.get('nextPageToken', None)
    if not page_token:
        break
```
